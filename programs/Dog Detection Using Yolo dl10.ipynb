```python
import cv2
from ultralytics import YOLO

model = YOLO("yolov8s.pt") 

image_path ="E:\download.jpeg"
image = cv2.imread(image_path)


results = model(image)

dog_found = False 

for result in results:
    boxes = result.boxes
    for box in boxes:
        cls_id = int(box.cls[0])
        conf = float(box.conf)
        if cls_id == 16:  # Dog class in COCO
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 100, 0), 2)
            cv2.putText(image, f"Dog {conf:.2f}", (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 100, 0), 2)
            dog_found = True

if not dog_found:
    blank = image.copy()
    cv2.putText(blank, "Dog Not Found", (50, 100),
                cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 0, 100), 3)
    cv2.imshow("Dog Detection", blank)
else:
    cv2.imshow("Dog Detection", image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
